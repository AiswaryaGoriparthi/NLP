```python
paragraph="""Dr. A.P.J. Abdul Kalam, often called the "Missile Man of India" and the “People’s President,” was known for his inspiring speeches that touched the hearts of millions, especially the youth. In his speeches, he emphasized the power of dreams, education, and hard work as the keys to success and national progress. He often said, “Dream, dream, dream. Dreams transform into thoughts and thoughts result in action,” urging young minds to have a vision and the courage to pursue it. Kalam’s speeches were filled with hope and positivity, motivating students to contribute to India’s development through innovation and integrity. He believed that every individual has a unique potential and that real education lies not just in acquiring knowledge, but in developing creativity, curiosity, and compassion. His address to students at IITs, schools, and universities often included personal stories from his own life—growing up in Rameswaram, working as a scientist at ISRO and DRDO, and serving as India’s President—all of which illustrated his humility and perseverance. Through his simple yet powerful words, Abdul Kalam inspired generations to think big, overcome challenges, and work selflessly for the betterment of society and the nation.
"""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
#Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=''.join(words)
```

```python
from nltk.stem import SnowballStemmer
SnowballStemmer=SnowballStemmer('english')
```

```python
#Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[SnowballStemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=''.join(words)
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
#Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=''.join(words)
```
