```python
!pip install nltk
words=["eating","eat","eaten","going","goes","gone","running","runs"]
```

```python
#PorterStemming
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in words:
  print(word+"----->"+stemming.stem(word))
```

```python
#PorterStemming
from nltk.stem import PorterStemmer
Port=PorterStemmer()
```

```python
for word in words:
  print(word+"------>"+Port.stem(word))
```

```python
#RegExpStemming
from nltk.stem import RegexpStemmer
Regexp=RegexpStemmer('ing$|s$|e$|able', min=4)
```

```python
Regexp.stem('eating')
```

```python
Regexp.stem('ableeatable')
```

```python
from nltk.stem import SnowballStemmer
snowball=SnowballStemmer('english')
```

```python
#SnowballStemming
for word in words:
  print(word+"--------->"+snowball.stem(word))
```

```python
Port.stem("fairly"),Port.stem("sportingly")
```
